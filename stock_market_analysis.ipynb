### Importing Numpy ,Pandas , Seaborn , Matplotlib , Pandas , Linear Regression and Time Delta

```python
# import 4 libraries
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
from datetime import timedelta
from sklearn.linear_model import LinearRegression

# creating the data frame
df = pd.read_excel('stock_data.xlsx')

# displaying the data frame
df

```

```python
# check for missing values
print(df.isnull().sum())
```

```python
# data cleaning and handle missing values
# Check for missing values
if df.isnull().values.any():
    print("Missing values found. Filling missing values with column means...")
    df.fillna(df.select_dtypes(include=[np.number]).mean(), inplace=True)
    print("Data cleaning has been successfully executed!\n")
else:
    print("No missing values found.")

# Display the DataFrame
print(df)
```

```python
#Removing Duplicates
# Check for duplicates
if df.duplicated().any():
    print("Duplicates found. Removing duplicates...")
    df.drop_duplicates(inplace=True)
    print("Data after removing duplicates:")
else:
    print("No duplicates found.")

# Display the DataFrame
print(df)

```

```python
# Convert the Date column to a datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Sort the DataFrame by the 'Date' column
# df.sort_values(by=['Date'], inplace=True)

# Reset the index to make it sequential
# df.reset_index(drop=True, inplace=True)

# Display the DataFrame
df.set_index("Date", inplace=True)
print(df)
```

## Visualizing the Data

# 1️⃣ Stock Closing Prices Over Time 📈
# 📌 Visualises the stock price trends over time for each stock.
# 📌 Helps identify patterns, peaks, and downward trends.

```python
#  1🏠 Plot stock closing prices over time

plt.figure(figsize=(12, 6))
for stock in df["Stock_Name"].unique():
    stock_data = df[df["Stock_Name"] == stock]
    plt.plot(stock_data.index, stock_data["Close"], label=stock)

plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("Stock Closing Prices Over Time")
plt.legend()
plt.grid()
plt.show()
```

# 2️⃣ Market Share Analysis 📊
# 📌 Shows the market share of different stocks based on their average closing prices.

```python
# 2🌟 Market Share Analysis
plt.figure(figsize=(8, 8))
market_share = df.groupby("Stock_Name")["Close"].mean()
plt.pie(market_share, labels=market_share.index, autopct="%1.1f%%", startangle=140, colors=sb.color_palette("muted"))
plt.title("Market Share Based on Average Closing Price")
plt.show()
```

# 3️⃣ Moving Average (Trend Analysis) 🔄
# 📌 Calculates a 7-day moving average to smooth out fluctuations and highlight trends.

```python
# 3📊 Moving Average (Trend Analysis)
df["Moving_Avg_7"] = df.groupby("Stock_Name")["Close"].transform(lambda x: x.rolling(window=7).mean())
plt.figure(figsize=(12, 6))
for stock in df["Stock_Name"].unique():
    stock_data = df[df["Stock_Name"] == stock]
    plt.plot(stock_data.index, stock_data["Moving_Avg_7"], label=f"{stock} 7-day MA")

plt.xlabel("Date")
plt.ylabel("Price")
plt.title("7-Day Moving Average of Closing Prices")
plt.legend()
plt.grid()
plt.show()
```

# 4️⃣ Heatmap: Correlation Matrix 🌡️
📌 Shows relationships between stock features (e.g., open price, close price, volume).

```python
# 4🌎 Heatmap: Correlation Matrix
plt.figure(figsize=(10, 6))
corr_matrix = df[["Open", "Close", "Volume", "Daily_Change"]].corr()
sb.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Stock Data")
plt.show()
```

# 5️⃣ Calculating the Profit and Loss 📊


```python
print(df["PL_Label"].value_counts())  # Check if "Loss" is even present
```

```python
# Randomly select 20% of the dataset and artificially reduce the closing price to create loss cases
df.loc[df.sample(frac=0.2).index, "Close"] = df["Open"] - np.random.uniform(1, 10, size=len(df))

# Calculate the profit or loss for each stock
df["Profit/Loss"] = df["Close"] - df["Open"]

# Label the stock as "Profit" if closing price is greater than or equal to opening price, otherwise label as "Loss"
df["PL_Label"] = np.where(df["Profit/Loss"] >= 0, "Profit", "Loss")
```

# 6️⃣ Scatter Plot: Profit vs Loss 💰📉
# 📌 Distinguishes profitable and loss-making stocks using a scatter plot.

```python
# 6📈 Scatter Plot: Profit vs Loss
df["Profit/Loss"] = df["Close"] - df["Open"]
df["PL_Label"] = np.where(df["Profit/Loss"] >= 0, "Profit", "Loss")
plt.figure(figsize=(10, 6))
sb.scatterplot(x=df["Open"], y=df["Close"], hue=df["PL_Label"], palette={"Profit": "green", "Loss": "red"})
plt.xlabel("Opening Price")
plt.ylabel("Closing Price")
plt.title("Scatter Plot: Profit vs Loss")
plt.grid()
plt.show()
```

# 7️⃣ Predictive Analysis: Next 30 Days Forecast 🔮📅
# 📌 Generates a future price trend based on historical data.

```python
# 7📈 Predictive Analysis (30-day forecast using Linear Regression)
df_reset = df.reset_index()
df_reset["Days"] = (df_reset["Date"] - df_reset["Date"].min()).dt.days

# Train model
model = LinearRegression()
X = df_reset[["Days"]]
y = df_reset["Close"]
model.fit(X, y)

# Generate future dates
future_days = np.array(range(df_reset["Days"].max() + 1, df_reset["Days"].max() + 31)).reshape(-1, 1)
future_dates = [df_reset["Date"].max() + timedelta(days=i) for i in range(1, 31)]
predictions = model.predict(future_days)

# Plot predictions
plt.figure(figsize=(12, 6))
plt.plot(df_reset["Date"], df_reset["Close"], label="Actual Data")
plt.plot(future_dates, predictions, label="Predicted", linestyle="dashed", color="red")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("Stock Price Prediction for Next 30 Days")
plt.legend()
plt.grid()
plt.show()
```
